## Read in the data
To just get all of the loading over with, I'm going to read the file in with csv.reader while doing some basic error-checking along the way based on the format of the data. Since I'm mainly interested in arrival and departure delays, I'm going to drop all flights that were cancelled. Also, any flights that for whatever reason have no arrival or departure delay (let me rephrase: that have a missing value; a 0 minute delay will be accepted) will be dropped. Other missing values will get filled in with some place holder value (either -999, or 'NA') so they will be easy to identify.

What follows is a listing of all the fields contained in the data.

Year is all 2008, month is 1-12, day of month is 1-31, day of week is 1-7 (Monday - Sunday). 

Distance is miles, all delays or time *amounts* are in number of minutes.

The departure/arrival time is a clock-time string, so '659' is 6:59 am, and '670' is an invalid time, but I'm going to convert these into minute-of-the-day, so '659' becomes $6 * 60 + 59 = 419$, just so the graphs are smoother.

```python
fields = ["year", "month", "day_of_month", "day_of_week", "dep_time",
          "sched_dep_time", "arr_time", "sched_arr_time", "carrier", "flight_num",
          "tail_num", "elapsed_time", "sched_elapsed_time", "air_time", "arr_delay",
          "dep_delay", "origin", "dest", "distance", "taxi_in",
          "taxi_out", "was_cancelled", "cancel_code", "was_diverted", "carrier_delay",
          "weather_delay", "NAS_delay", "security_delay", "late_aircraft_delay"]
non_numeric_fields = set([fields.index(f) for f in fields 
                          if f in ("carrier", "tail_num", "origin", "dest", "cancel_code")])
clock_fields = set([fields.index(f) for f in fields 
                    if f in ("sched_dep_time", "dep_time", "sched_arr_time", "arr_time")])
drop_if_true_fields = set([fields.index(f) for f in fields if f in ("cancelled")])
drop_if_na_fields = set([fields.index(f) for f in fields if f in ("arr_delay", "dep_delay")])
print(clock_fields)
```

```python
data = []
with open("../data/airplane.thin.2008.csv") as fh:
    for line in csv.reader(fh):
        ## skip row if cancelled, or missing arrival/departure delay (etc)
        skip = False
        for field in drop_if_true_fields:
            if line[field] == '1':
                skip = True
        for field in drop_if_na_fields:
            if line[field] in ('NA', ''):
                skip = True
        if skip:
            continue
        ## Parse the data as necessary
        parsed = []
        for idx, variable in enumerate(line):
            if idx in non_numeric_fields:
                if variable == '':
                    variable = 'NA'
                parsed.append(variable)
            elif idx in clock_fields:
                try:
                    hrs, mins = int(variable[:-2]), int(variable[-2:])
                    parsed.append(hrs * 60 + mins)
                except ValueError as e:
                    parsed.append(-999)
            else:
                try:
                    parsed.append(float(variable))
                except ValueError as e:
                    parsed.append(-999)
        data.append(parsed)
print(len(data), "\n", data[0])
```

I think 100,000 seems to work well; this should give enough data to play around with and ask interesting questions, without having too much that things run slowly.

I've tried abstracting these plotting techniques into simple functions, but we'll see how well they pan out.

```python
def boxplot_data(df, column, ax):
    ax.boxplot(df[column])
    x = np.random.normal(1, 0.08, size=len(df[column]))
    ax.plot(x, df[column], 'r.', alpha=0.2)
    ax.set_xlabel(column)

def histplot_data(df, column, ax, xmin=None, xmax=None, nbins=30):
    MIN, MAX = min(df[column]), max(df[column])
    if xmin is None:
        xmin = MIN
    if xmax is None:
        xmax = MAX
    width = (xmax - xmin)/nbins
    normal_ys, normal_bins = np.histogram(df[column], bins=nbins)
    normal_bin_centers = 0.5*(normal_bins[1:]+normal_bins[:-1])
    ax.bar(normal_bin_centers, normal_ys, width=width, color=next(colors))
    ax.set_xlim(xmin, xmax)
    ax.set_xlabel(column)
```

## Interactions:
***
Alright, so we've looked at a handful of variables and how they relate to average departure delay (and average departure delays more than 30 minutes). Delta has the shortest delay time; Wednesday has the shortest delay time; September has the shortest delay time; 6am has the shortest delay time. Can we plain and simple say then that people ought to book their trips on Wednesday mornings in September with Delta? And definitely *not* book trips with JetBlue on Friday evenings in August?
```python
carriers = {'B6' : 'JetBlue', 'UA' : 'United Airlines', 'DL' : 'Delta'}
fig, ax = plt.subplots(ncols=3, figsize=figsize(12, 5))
max_delay = 0
for idx, discriminant in enumerate(set(df["carrier"])):
    inner = df[df.carrier == discriminant]
    inner30 = df30[df30.carrier == discriminant]
    keys, all_delays, half_delays = [], [], []
    key_names = ["", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
    for key, sub_inner in inner.groupby(by="day_of_week"):
        keys.append(key)
        all_delays.append(np.mean(sub_inner["dep_delay"]))
    for key, sub_inner in inner30.groupby(by="day_of_week"):
        half_delay = np.mean(sub_inner["dep_delay"])
        half_delays.append(half_delay)
        max_delay = max([max_delay, half_delay])
    ax[idx].bar(range(len(keys)), half_delays, align='center', alpha=0.5, color=gr)
    ax[idx].bar(range(len(keys)), all_delays, align='center', alpha=0.5, color=ltbl)
    ax[idx].set_xticklabels(key_names, rotation=45, ha='right')
    ax[idx].set_title(carriers[discriminant])
    
## Trying to neaten up the graphs a little...
for i in range(3):
    plt.sca(ax[i])
    plt.ylim(0, max_delay + max_delay/10)
    plt.xlabel("Day of Week")
plt.sca(ax[0]) 
plt.legend(["30+ min late", "all delays"], loc="upper center")
plt.ylabel("Average departure delay")
plt.subplots_adjust(wspace=0.2)
```

The graphs look mostly as we should have expected: shortest average delay across carriers at 6 in the morning, and it increases throughout the day, then starts tapering off again after 8pm.
